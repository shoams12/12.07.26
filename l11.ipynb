```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

y = np.array([10, 12, 20, 18])
y_pred = np.array([11, 13, 18, 18])
y_mean = np.mean(y)
#Compute r2 by hand (no sklearn) for this tiny dataset.
sse = sum((y - y_pred)**2)
sst = sum((y - y_mean)**2)
R2 = 1 - sse / sst
print(R2)
print(r2_score(y, y_pred))
```

```python
X = np.array([
    [2, 1, 3], [4, 2, 1], [6, 3, 5], [8, 1, 4],
    [3, 5, 2], [5, 4, 6], [7, 2, 3], [9, 5, 1],
    [1, 3, 4], [10, 4, 5]
])

Y = np.array([18, 20, 36, 40, 15, 32, 33, 30, 17, 43])
```

```python
model = LinearRegression()
model.fit(X, Y)
print(model.intercept_)
print(model.coef_)
```

```python
y_pred = model.predict(X)
mse = mean_squared_error(Y, y_pred)
print(f'MSE: {mse}')
```

```python
R2 = r2_score(Y, y_pred)
print(f'R2: {R2}')
```
